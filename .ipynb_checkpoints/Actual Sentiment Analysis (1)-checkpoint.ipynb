```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
from nltk.tokenize import TweetTokenizer
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
import re
import joblib
import seaborn as sns
```

```python

# English Sentiment Analysis

# Load the English dataset
df_en = pd.read_csv('English.csv', header=None, encoding="ISO-8859-1")
df_en.columns = ['target', 'text']

# Convert target values from 4 to 1 for positive sentiment
df_en['target'] = df_en['target'].replace(4, 1)

# Define custom English word tokenizer (using TweetTokenizer)
tokenizer_en = TweetTokenizer()

# Define English stop words
stop_words_en = set(stopwords.words('english'))

# English Text preprocessing function
def preprocess_text_en(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|@[^\s]+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    tokens = tokenizer_en.tokenize(text)
    tokens = [token for token in tokens if token not in stop_words_en]
    return ' '.join(tokens)
```

```python
# Apply text preprocessing to the 'text' column
df_en['text'] = df_en['text'].apply(preprocess_text_en)

# Split the English dataset into training and testing sets
X_train_en, X_test_en, y_train_en, y_test_en = train_test_split(df_en['text'], df_en['target'], test_size=0.2, random_state=42)

# Create a TfidfVectorizer for English
tfidf_vectorizer_en = TfidfVectorizer(max_features=5000)

# Transform the English text data
X_train_tfidf_en = tfidf_vectorizer_en.fit_transform(X_train_en)
X_test_tfidf_en = tfidf_vectorizer_en.transform(X_test_en)

# Train the Logistic Regression model for English
logreg_model_en = LogisticRegression()
logreg_model_en.fit(X_train_tfidf_en, y_train_en)

# Evaluate English model
y_pred_en = logreg_model_en.predict(X_test_tfidf_en)
accuracy_en = accuracy_score(y_test_en, y_pred_en)

# Print English model accuracy
print(f"English Model Accuracy: {accuracy_en:.2f}")
```

```python
# Print English classification report
print("\nEnglish Classification Report:")
print(classification_report(y_test_en, y_pred_en))

# Plot English confusion matrix
conf_matrix_en = confusion_matrix(y_test_en, y_pred_en)
plt.imshow(conf_matrix_en, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('English Confusion Matrix')
plt.colorbar()
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
# Plot English ROC curve
y_prob_en = logreg_model_en.predict_proba(X_test_tfidf_en)[:, 1]
fpr_en, tpr_en, _ = roc_curve(y_test_en, y_prob_en)
roc_auc_en = auc(fpr_en, tpr_en)
plt.figure()
plt.plot(fpr_en, tpr_en, color='darkorange', lw=2, label=f'English ROC curve (area = {roc_auc_en:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('English Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
# Exploratory Data Analysis (EDA) for English dataset
print("English Dataset Summary:")
print(df_en.info())
print("\nEnglish Dataset Description:")
df_en['sentiment'] = df_en['target'].map({0: 'negative', 1: 'positive'})
print(df_en.describe())
print("\nDistribution of Sentiment Labels in English Dataset:")
sns.countplot(x='sentiment', data=df_en)  # Use 'sentiment' instead of 'target'
plt.title('Distribution of Sentiment Labels in English Dataset')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
# Arabic Sentiment Analysis
# Load the Arabic dataset
df_ar = pd.read_csv('Arabic.csv', names=['target', 'text'], header=None)

# Arabic Text preprocessing
stop_words_ar = set(stopwords.words('arabic'))
tokenizer_ar = RegexpTokenizer(r'\b\w+\b')  # Custom Arabic word tokenizer

def preprocess_text_ar(text):
    stop_words_ar = set(stopwords.words('arabic'))
    tokenizer_ar = RegexpTokenizer(r'\w+')
    words_ar = tokenizer_ar.tokenize(text)
    filtered_words_ar = [word for word in words_ar if word.isalpha() and word not in stop_words_ar]
    return ' '.join(filtered_words_ar)

df_ar['processed_text'] = df_ar['text'].apply(preprocess_text_ar)
```

```python
# Split the Arabic dataset into training and testing sets
train_data_ar, test_data_ar, train_labels_ar, test_labels_ar = train_test_split(df_ar['processed_text'], df_ar['target'], test_size=0.2, random_state=42)

# Convert multiclass labels to binary (positive class: 4, negative class: others)
train_labels_binary_ar = np.where(train_labels_ar == 4, 1, 0)
test_labels_binary_ar = np.where(test_labels_ar == 4, 1, 0)

# Create a TfidfVectorizer for Arabic
vectorizer_ar = TfidfVectorizer()
train_tfidf_ar = vectorizer_ar.fit_transform(train_data_ar)

# Train the SVM model for Arabic
svm_model_ar = SVC(C=1, kernel='linear', gamma='auto', probability=True)
svm_model_ar.fit(train_tfidf_ar, train_labels_binary_ar)
# Transform Arabic test data
test_tfidf_ar = vectorizer_ar.transform(test_data_ar)

# Make predictions on the Arabic test set
predictions_ar = svm_model_ar.predict(test_tfidf_ar)

# Evaluate the Arabic model
accuracy_ar = accuracy_score(test_labels_binary_ar, predictions_ar)
```

```python
print(f"Arabic Model Accuracy: {accuracy_ar:.2f}")

print("\nArabic Classification Report:")
print(classification_report(test_labels_binary_ar, predictions_ar))

# Plotting Arabic confusion matrix
conf_matrix_ar = confusion_matrix(test_labels_binary_ar, predictions_ar)
plt.imshow(conf_matrix_ar, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Arabic Confusion Matrix')
plt.colorbar()
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
# Plotting Arabic ROC curve
y_score_ar = svm_model_ar.decision_function(test_tfidf_ar)
fpr_ar, tpr_ar, _ = roc_curve(test_labels_binary_ar, y_score_ar)
roc_auc_ar = auc(fpr_ar, tpr_ar)
plt.figure()
plt.plot(fpr_ar, tpr_ar, color='darkorange', lw=2, label=f'Arabic ROC curve (area = {roc_auc_ar:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Arabic Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
# Map sentiment labels to 'negative', 'neutral', and 'positive'
df_ar['sentiment'] = df_ar['target'].map({0: 'negative', 4: 'positive'})

# Exploratory Data Analysis (EDA) for Arabic dataset
print("Arabic Dataset Summary:")
print(df_ar.info())
print("\nArabic Dataset Description:")
print(df_ar.describe())
print("\nDistribution of Sentiment Labels in Arabic Dataset:")
sns.countplot(x='sentiment', data=df_ar)  # Use 'sentiment' instead of 'target'
plt.title('Distribution of Sentiment Labels in Arabic Dataset')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python

# Load English and Arabic sentiment analysis models
logreg_model_en = joblib.load('english_sentiment_model.pkl')
tfidf_vectorizer_en = joblib.load('english_tfidf_vectorizer.pkl')

svm_model_ar = joblib.load('arabic_sentiment_model.pkl')
vectorizer_ar = joblib.load('arabic_tfidf_vectorizer.pkl')

# Define the sentiment analysis function
def analyze_sentiment(text):
    lang = 'en' if re.search('[a-zA-Z]', text) else 'ar'
    if lang == 'en':
        processed_text = preprocess_text_en(text)
        tfidf_text = tfidf_vectorizer_en.transform([processed_text])
        prediction = logreg_model_en.predict(tfidf_text)
    elif lang == 'ar':
        processed_text = preprocess_text_ar(text)
        tfidf_text = vectorizer_ar.transform([processed_text])
        prediction = svm_model_ar.predict(tfidf_text)
    return prediction[0]

# Function to preprocess English text
def preprocess_text_en(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|@[^\s]+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    tokens = tokenizer_en.tokenize(text)
    tokens = [token for token in tokens if token not in stop_words_en]
    return ' '.join(tokens)

# Function to preprocess Arabic text
def preprocess_text_ar(text):
    stop_words_ar = set(stopwords.words('arabic'))
    tokenizer_ar = RegexpTokenizer(r'\w+')
    words_ar = tokenizer_ar.tokenize(text)
    filtered_words_ar = [word for word in words_ar if word.isalpha() and word not in stop_words_ar]
    return ' '.join(filtered_words_ar)

# Read the CSV file containing mixed language statements
df = pd.read_csv('Test_sentiment.csv')

# Initialize lists to store positive and negative statements
positive_statements = []
negative_statements = []

# Analyze sentiments for each statement in the DataFrame
for index, row in df.iterrows():
    prediction = analyze_sentiment(row['text'])
    if prediction == 1:
        positive_statements.append(row['text'])
    else:
        negative_statements.append(row['text'])

# Print positive and negative statements
print("Positive Statements:")
for statement in positive_statements:
    print("- ", statement)
print("\nNegative Statements:")
for statement in negative_statements:
    print("- ", statement)
```
